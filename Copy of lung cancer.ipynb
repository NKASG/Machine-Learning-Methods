```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn import model_selection
from sklearn import linear_model
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
data =pd.read_csv('/content/sample_data/survey lung cancer.csv', header = 0 , delimiter = ",")
```

```python
data
```

```python
data.drop(labels = 'LUNG_CANCER' , axis = 1 , inplace = True)
data.drop(labels = 'GENDER' , axis = 1 , inplace = True)
```

```python
comp = data['COUGHING']
```

```python
data.drop(labels = 'COUGHING' , axis = 1 , inplace = True)
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(data,comp, test_size=0.2,train_size=0.8)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
model = linear_model.LinearRegression()
```

```python
X_train_1 = X_train.values
X_test_1 = X_test.values
y_train_1 = y_train.values
y_test_1 = y_test.values
```
