```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/commodity_2000_2022.csv')
```

```python
data.head()
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data.Symbol = le.fit_transform(data.Symbol)
data.Date = le.fit_transform(data.Date)
```

```python
label_1 = data['Open']
```

```python
label_2 = data['High']
label_3 = data['Low']
label_4 = data['Close']
```

```python
label = pd.concat([label_1,label_2,label_3,label_4],axis=1)
```

```python
label
```

```python
data.drop(labels= ['Open', 'High' , 'Low' , 'Close' ] , axis = 1, inplace=True)
```

```python
import sklearn
from sklearn import model_selection
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(data,label, test_size=0.2,train_size=0.8)
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
```

```python
X_train_1 = X_train.values
X_test_1 = X_test.values
y_train_1 = y_train.values
y_test_1 = y_test.values
```
